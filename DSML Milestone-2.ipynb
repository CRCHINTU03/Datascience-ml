<h1>2. Making data Set </h1>

In the three data sets we have two common attributes 'date' and 'location_key'. First we will fetch one row from each dataset and check whether the entry exists in the other datasets on same date and same location.
Now Based on the attributes date and location_key we will merge the three datasets.

```python
#checking for an entry on date 2020-09-28 at location AR in the dataset by-sex
bysex.loc[(bysex['date'] == '2020-09-28') & (bysex['location_key'] == 'AR')]
```

```python
#checking for an entry on date 2020-09-28 at location AR in the dataset epidiamology
epid.loc[(epid['date'] == '2020-09-28') & (epid['location_key'] == 'AR')]
```

```python
#checking for an entry on date 2020-09-28 at location AR in the dataset hospitalizations
hosp.loc[(hosp['date'] == '2020-09-28') & (hosp['location_key'] == 'AR')]
```

```python
#mergind datasets hospitalizations and by-sex
merge1=pd.merge(hosp, bysex, on=['date','location_key'])
```

```python
#merging the datasets merge1 and epidiamology
final_merge=pd.merge(merge1, epid, on=['date','location_key'])
```

# Heatmap
Heatmap is a good way to quickly check correlations among columns by visualizing the correlation matrix using heatmap.

```python
mask = np.zeros_like(df.corr())
mask[np.triu_indices_from(mask)] = True

sns.heatmap(df.corr(), cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=2)
plt.rcParams['figure.figsize'] = (18, 18)
plt.yticks(rotation=0)
plt.xticks(rotation=60)

plt.savefig('correlation_heatmap.png',dpi=500,bbox_inches='tight')
```

##  Normalize Data

```python
#Normalize the data in all the columns from new_hospitalized_patients to cumulative tested
data_norm = scale(df.loc[:,'new_hospitalized_patients':'cumulative_tested'])
```

```python
#describe the data after normalization
pd.DataFrame(data_norm).describe().transpose()
```

```python
#finding the correlation among the normalized data
pd.DataFrame(data_norm).corr()
```

```python
# Selecting the number of components
pca1 = PCA(n_components=36)

# Fit the PCA model
pca1.fit(data_norm)
```

```python
#The amount of variance that each PC explains
var = pca1.explained_variance_ratio_
print(var)
```

```python
#Cumulative Variance explains
var1 = np.cumsum(np.round(pca1.explained_variance_ratio_, decimals=4)*100)
print(var1)
```

```python
var1 = pd.DataFrame(var1, index=np.arange(1,37))
plt.plot(var1,color='blue')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.savefig('scree_plot.png',dpi=100,bbox_inches='tight')
```
